# Using ISeeU for ICU mortality prediction

```python
from iseeu import ISeeU
import numpy as np
```

```python
ISeeU.__version__
```

```python
patient = np.load('patient.npy')
```

```python
patient.shape
```

```python
predictor = ISeeU()
```

```python
predictor.visualize_patient_scores(patient_tensor=patient)
```

```python
prediction, scores = predictor.predict(patient)
```

```python
print(prediction)
print(scores.shape)
print(patient.shape)
```

```python
predictor.visualize_patient_scores(patient_tensor=patient, importance_scores=scores)
```

```python
predictor.visualize_evidence(importance_scores=scores)
```
